```python
# Standard libraries
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt

# Project modules
from src.data_io import load_all
from src.pipeline import build_ratings
from src.elo import calculate_new_elo, expected_position_diff
```

```python
# Load CSVs from data folder
data_drivers, data_races, data_results = load_all()
```

```python
# Compute Elo ratings
Ratings_History, driver_index_map, sorted_race_ids = build_ratings(
    data_drivers, data_races, data_results
)
```

```python
def get_driver_index(driver_id):
    """Return row index of a driver in Ratings_History."""
    return driver_index_map[driver_id]

def get_driver_name(driver_id):
    """Return driverRef (or name) from drivers DataFrame."""
    row = data_drivers[data_drivers['driverId'] == driver_id]
    return row['driverRef'].values[0]
```

```python
# Compare multiple drivers by typing in their IDs
driver_ids = [1, 4, 7]  # replace with actual driver IDs
for d_id in driver_ids:
    idx = get_driver_index(d_id)
    plt.plot(Ratings_History[idx, :], label=get_driver_name(d_id))

plt.title("Driver Elo Comparison")
plt.xlabel("Race Number")
plt.ylabel("Elo Rating")
plt.legend()
plt.show()
```
